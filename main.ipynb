# Visualizing model output

```python
import matplotlib.pyplot as plt
import numpy as np
import rasterio
import matplotlib

NO_DATA = -9999
NO_DATA_FLOAT = 0.0001
PERCENTILES = (0.1, 99.9)
```

```python
# reads & returns raster image
def load_raster(path, crop=None):
    with rasterio.open(path) as src:
        img = src.read()

        # load first 6 bands
        img = img[:6]

        img = np.where(img == NO_DATA, NO_DATA_FLOAT, img)
        if crop:
            img = img[:, -crop[0]:, -crop[1]:]
    return np.squeeze(img)
```

```python
# normalizes raster image to RGB

def enhance_raster_for_visualization(raster, ref_img=None):
    if ref_img is None:
        ref_img = raster
    channels = []
    for channel in range(raster.shape[0]):
        valid_mask = np.ones_like(ref_img[channel], dtype=bool)
        valid_mask[ref_img[channel] == NO_DATA_FLOAT] = False
        mins, maxs = np.percentile(ref_img[channel][valid_mask], PERCENTILES)
        normalized_raster = (raster[channel] - mins) / (maxs - mins)
        normalized_raster[~valid_mask] = 0
        clipped = np.clip(normalized_raster, 0, 1)
        channels.append(clipped)
    clipped = np.stack(channels)
    channels_last = np.moveaxis(clipped, 0, -1)[..., :3]
    rgb = channels_last[..., ::-1]
    return rgb

```

```python
#visualize model inference

norm = matplotlib.colors.Normalize(vmin=0, vmax=2)

fig, ax = plt.subplots(1, 4, figsize=(15, 10))
originalImage = enhance_raster_for_visualization(load_raster("inference-images/subsetted_512x512_HLS.S30.T11SMU.2020299.v1.4_merged.tif"))
groundTruth = load_raster("subsetted_512x512_HLS.S30.T11SMU.2020299.v1.4.mask.tif")
modelInference = load_raster("results/finetuning-first/subsetted_512x512_HLS.S30.T11SMU.2020299.v1.4_merged_pred.tif")
checkpointModelInference = load_raster("results/Prithvi-100M-burn-scar/subsetted_512x512_HLS.S30.T11SMU.2020299.v1.4_merged_pred.tif")

for subplot in ax:
    subplot.axis('off')

ax[0].imshow(originalImage)
ax[0].set_title("Original Image")
ax[1].imshow(groundTruth, norm=norm, cmap="jet")
ax[1].set_title("Ground Truth")
ax[2].imshow(modelInference, norm=norm, cmap="jet")
ax[2].set_title("80-Iteration Model Inference")
ax[3].imshow(checkpointModelInference, norm=norm, cmap="jet")
ax[3].set_title("Checkpoint Model Inference")
#ax[3].imshow(originalImage)
#ax[3].imshow(load_raster(modelInference), norm=norm, cmap="jet", alpha=0.3)
#ax[3].set_title("Overlay")
```
